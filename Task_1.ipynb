```python
from lib import bn256
import random
import copy
```

```python
R = [random.randrange(2, bn256.order) for _ in range(3)]
r1, r2, r3 = R
r1, r2, r3
```

```python
G1 = bn256.curve_G
G2 = bn256.twist_G
G1, G2
```

```python
T13 = [G1.scalar_mul(r) for r in R]
T1, T2, T3 = T13
T1, T2, T3
```

```python
type(T1)
```

```python
T3_inv = copy.copy(T3)
T3_inv.y = T3_inv.y.additive_inverse()
T3_inv
```

```python
TP = T1.add(T2).add(T3_inv)
TP
```

```python
T46 = [G2.scalar_mul(r) for r in R]
T4, T5, T6 = T46
T4, T5, T6
```

```python
type(T4), type(T4.y)
```

```python
T6_inv = copy.copy(T6)
T6_inv.negate()
T6_inv
```

```python
TQ = T4.add(T5).add(T6)
TQ
```

```python
tf = r1 + r2 - r3
tf
```
