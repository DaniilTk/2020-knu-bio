```python
import random
import bn256
```

```python
r1 = random.randrange(2,bn256.order)
r2 = random.randrange(2,bn256.order)
r3 = random.randrange(2,bn256.order)
```

```python
r1,r2,r3
```

```python
G1 = bn256.curve_G
G1
```

```python
G2 = bn256.twist_G
G2
```

```python
T1 = G1.scalar_mul(r1)
T1
```

```python
T2 = G1.scalar_mul(r2)
T2
```

```python
T3 = G1.scalar_mul(r3)
T3
```

```python
object_1 = bn256.gfp_1(T3.x.value())
object_2 = bn256.gfp_1(-1*T3.y.value())
```

```python
print(object_1, type(object_1))
print(object_2, type(object_2))
```

```python
rev_T3 = bn256.curve_point(T3.x, object_2)
rev_T3
```

```python
# TP = T1 + T2 - T3
TP = (T1.add(T2)).add(rev_T3)
TP
```

```python
T4 = G2.scalar_mul(r1)
T4
```

```python
T5 = G2.scalar_mul(r2)
T5
```

```python
T6 = G2.scalar_mul(r3)
T6
```

```python
# TQ = T4 + T5 - T6
TQ = T4.add(T5).add(T6.scalar_mul(bn256.order - 1))
TQ
```

```python
TF = r1 + r2 - r3
TF
```

```python
# E1 = pairing(G1, G2) ^ (tf ^ 2)
E1 =  bn256.optimal_ate(G2, G1).exp(TF**2) 
E1
```

```python
# E2 = pairing(TQ, TP)
E2 = bn256.optimal_ate(TQ, TP)
E2
```

```python
E1 == E2
```
